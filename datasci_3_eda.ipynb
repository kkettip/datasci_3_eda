# Central tendency (mean, median, mode)

```python
#For Total Telehealth Users; column : Total_Bene_Telehealth_Integer_value
mean1 = new_df['Total_Bene_Telehealth_Integer_value'].mean()
median1 = new_df['Total_Bene_Telehealth_Integer_value'].median()
mode1 = new_df['Total_Bene_Telehealth_Integer_value'].mode().values[0]
print(f'Mean1: {mean1}')
print(f'Median1: {median1}')
print(f'Mode1: {mode1}')
```

```python
#For Total Telehealth Eligible Users; column : Total_Bene_TH_Elig_Integer_value

mean2 = new_df['Total_Bene_TH_Elig_Integer_value'].mean()
median2 = new_df['Total_Bene_TH_Elig_Integer_value'].median()
mode2 = new_df['Total_Bene_TH_Elig_Integer_value'].mode().values[0]
print(f'Mean2: {mean2}')
print(f'Median2: {median2}')
print(f'Mode2: {mode2}')
```

```python
#For Percentage of Medicare Users with a Telehealth Service; column : Pct_Telehealth

mean3 = new_df['Pct_Telehealth'].mean()
median3 = new_df['Pct_Telehealth'].median()
mode3 = new_df['Pct_Telehealth'].mode().values[0]
print(f'Mean3: {mean3}')
print(f'Median3: {median3}')
print(f'Mode3: {mode3}')
```

# Measures of spread (range, variance, standard deviation, IQR)

```python
#For Percentage of Medicare Users with a Telehealth Service; column : Pct_Telehealth

range1 = new_df['Pct_Telehealth'].max() - new_df['Pct_Telehealth'].min()
variance1 = new_df['Pct_Telehealth'].var()
std_deviation1 = new_df['Pct_Telehealth'].std()
iqr1 = new_df['Pct_Telehealth'].quantile(0.75) - new_df['Pct_Telehealth'].quantile(0.25)
print(f'Range1: {range1}')
print(f'Variance1: {variance1}')
print(f'Standard deviation1: {std_deviation1}')
print(f'IQR1: {iqr1}')
```

```python
#For Total Telehealth Users; column : Total_Bene_Telehealth_Integer_value

range2 = new_df['Total_Bene_Telehealth_Integer_value'].max() - new_df['Total_Bene_Telehealth_Integer_value'].min()
variance2 = new_df['Total_Bene_Telehealth_Integer_value'].var()
std_deviation2 = new_df['Total_Bene_Telehealth_Integer_value'].std()
iqr2 = new_df['Total_Bene_Telehealth_Integer_value'].quantile(0.75) - new_df['Total_Bene_Telehealth_Integer_value'].quantile(0.25)
print(f'Range2: {range2}')
print(f'Variance2: {variance2}')
print(f'Standard deviation2: {std_deviation2}')
print(f'IQR: {iqr2}')
```

```python
#For Total Telehealth Eligible Users; column : Total_Bene_TH_Elig_Integer_value

range3 = new_df['Total_Bene_TH_Elig_Integer_value'].max() - new_df['Total_Bene_TH_Elig_Integer_value'].min()
variance3 = new_df['Total_Bene_TH_Elig_Integer_value'].var()
std_deviation3 = new_df['Total_Bene_TH_Elig_Integer_value'].std()
iqr3 = new_df['Total_Bene_TH_Elig_Integer_value'].quantile(0.75) - new_df['Total_Bene_TH_Elig_Integer_value'].quantile(0.25)
print(f'Range3: {range3}')
print(f'Variance3: {variance3}')
print(f'Standard deviation3: {std_deviation3}')
print(f'IQR3: {iqr3}')
```

# Handling Outliers

```python
Q1 = df['Pct_Telehealth'].quantile(0.25)
Q3 = df['Pct_Telehealth'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound)
print(upper_bound)
```

```python
outliers = df[(df['Pct_Telehealth'] < lower_bound) | (df['Pct_Telehealth'] > upper_bound)]
print(outliers)
```

```python
outliers.count()
```

A total of 22 outliers for each of the column categories are identified. I decided to retain the outliers because it may represent valid cases with valuable information that are not due to input errors.
